# Exploratory Data Analysis

# Gender

```python
# ploting a bar graph for Gender and it's count 

ax=sns.countplot(x='Gender',data=ds)
 
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Ploting a bar chart for Gender VS Amount 

sales_gen = ds.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

For the above graph we can conclude that most of the buyers are Female and even purchasing power is greater than men

# Age

```python
ax= sns.countplot(data=ds , x='Age_group', hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount Vs Age Group

sales_age=ds.groupby(['Age_group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age_group', y='Amount', data=sales_age)
```

From the above grapgh we conclude that most of the buyers(Female) are between 26-35 age group

# State

```python
# Total amount/sales from top 15 states 

sales_state= ds.groupby(['State'], as_index=False).sum().sort_values(by='Amount', ascending=False).head(15)

sns.set(rc={'figure.figsize':(25,15)})
sns.barplot(x='State', y='Amount', data=sales_state)
```

From the above graph we can see that most orders & total sales/Amount are from Uttar pradesh,Maharashtra,Karnataka respectively

# Marital Status 

```python
ax= sns.countplot(data=ds , x='Marital_Status')

sns.set(rc={'figure.figsize':(10,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=ds.groupby(['Marital_Status','Gender'], as_index=False).sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales_state, x='Marital_Status',y='Amount', hue='Gender')
```

From the above graph we can see that most of the buyers are married(women) and have high purchasing power

# Occupation 

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=ds , x='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state= ds.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation', y='Amount', data=sales_state)
```

For the above graph we can see that most of the buyers are from IT Sector,Healthcare & Aviation

# Product Category

```python
# Countplot

sns.set(rc={'figure.figsize':(25,10)})
ax=sns.countplot(data=ds, x='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# First 10 products from Product_Category

sales_state=ds.groupby(['Product_Category'], as_index=False).sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data=sales_state, x='Product_Category', y='Amount')
```

From the above graph we see that most of the sold products are from Food,Clothing&Apparel and Electronic gadgets

```python
sales_state=ds.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data=sales_state, x='Product_ID', y='Orders')
```

```python
# Top 10 most sold products (same as above)

fig1 , ax1=plt.subplots(figsize=(12,7))
ds.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
